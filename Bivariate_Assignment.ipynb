```python
# Test the Analysis of Variance between etest_p and mba_p at signifance
# level 5%.(Make decision using Hypothesis Testing)
import scipy.stats as stats
stats.f_oneway(dataset['etest_p'],dataset['mba_p'])
#P value is far greater than 0.05 so there is a significant diff between etest_p and mba_p
```

```python
# Test the similarity between the degree_t(Sci&Tech) and specialisa
# tion(Mkt&HR) with respect to salary at significance level of 5%.(Make
#  decision using Hypothesis Testing)
dataset_sci = dataset[dataset.degree_t=='Sci&Tech']
dataset_hr = dataset[dataset.specialisation=='Mkt&HR']
stats.f_oneway(dataset_sci['salary'],dataset_hr['salary']
#p value is 0.007 which is less than 0.05 accept level mentioned so accept null hypothesis
#there is no significant difference between students salary who studieed in  sci and tech degree 
               #versus students salary who studied in market and hr department
```

```python
# 9)Convert the normal distribution to standard normal distribution for
#  salary column
#when trying standard distribution for salary i was gettting two curves one for placed and one for not placed.
#so i filtered only placed students to see valid standard distribution
def stndgraph(dataset):
    import seaborn as sns
    
    mean = dataset.mean()
    std = dataset.std()
    
    values = [i  for i in dataset]
    z_score = [((j-mean)/std)for j in values]
    sns.distplot(z_score,kde=True)
    sum(z_score)/len(z_score)
```

```python
dataset_placed = dataset[dataset.status=='Placed']
stndgraph(dataset_placed['salary'])
```

```python
# What is the probability Density Function of the salary range from
#  700000 to 900000

def get_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    sample = dataset
    sample_mean = sample.mean()
    sample_std = sample.std()
    print('Mean=',(sample_mean))
    print('Standard deviation=',(sample_std))
    dist = norm(sample_mean,sample_std)
    values = [value for value in range(startrange,endrange)]
    probabilities = [dist.pdf(value)for value in values]
    prob = sum(probabilities)
    print('probability is ',prob)
    return prob
```

```python
get_probability(dataset['salary'],700000,900000)
#probability of getting salary from 7 lakh to 9 lakh is very minimal less than 1 percent
```
